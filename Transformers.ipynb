```python
import datasets
import huggingface_hub
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import transformers
```

```python
from transformers import pipeline
```

Text Classification

```python
model=pipeline('text-classification')
```

```python
text='''The Little Prince" is a philosophical tale that follows the journey of a young prince who travels from planet to planet, meeting various inhabitants and learning about human nature and life. The story begins with a pilot stranded in the Sahara desert who meets the Little Prince, a mysterious child from another world. As they repair the pilot's plane, the Little Prince recounts his adventures on different asteroids, each inhabited by a single adult who embodies various human flaws, such as a king, a vain man, a drunkard, a businessman, a lamplighter, and a geographer.

The Little Prince's home is a small asteroid named B-612, where he cares for three volcanoes and a unique rose. His relationship with the rose is complex; he loves her, but her vanity and demands cause him pain, prompting him to leave and explore the universe. Throughout his journey, the Little Prince learns important lessons about love, responsibility, and the essence of human connections.

The story emphasizes the importance of looking beyond the superficial to understand the true nature of things, symbolized by the Little Prince's repeated observation that "what is essential is invisible to the eye." The book concludes with the pilot reflecting on the lessons he learned from the Little Prince and contemplating the mysteries of life and love.'''
```

```python
model(text)
```

Named Entity Recognition

```python
model=pipeline('ner')
```

```python
res=model(text,aggregation_strategy='simple')
```

```python
pd.DataFrame(res)
```

Question Answer

```python
model=pipeline('question-answering')
```

```python
ques1='Where do the little prince travel?'
ques2="What is the little prince's home?"
```

```python
res1=model(ques1,text)
res2=model(ques2,text)
```

```python
res1
```

```python
res2
```

Text Summarization

```python
model=pipeline('summarization')
```

```python
res=model(text,max_length=100,clean_up_tokenization_spaces=True)
```

```python
res
```

Translation

```python
model=pipeline('translation_en_to_hi',model='Helsinki-NLP/opus-mt-en-hi')
```

```python
res=model(text,clean_up_tokenization_spaces=True,min_length=100)
```
